```python
print(df3_merge.info())
```

```python
# Calculate leverage
df3_merge['Leverage'] = df3_merge['Size USD'] / df3_merge['Start Position']

# View the new DataFrame
print(df3_merge.head())
print(df3_merge.info())
```

```python
# Filter out zero or near-zero Start Positions to avoid infinite leverage
threshold = 1e-6  # Small positive threshold to avoid tiny denominators
df4_filtered = df3_merge[df3_merge['Start Position'] > threshold].copy()

# Recalculate leverage on cleaned data
df4_filtered['Leverage'] = df4_filtered['Size USD'] / df4_filtered['Start Position']

print(df4_filtered.head())
```

```python
df4_filtered.to_csv('data_filtered.csv')
```

```python
files.download('data_filtered.csv')
```

```python
profitability_stats = df4_filtered.groupby('classification')['Closed PnL'].agg(['mean', 'median', 'count']).reset_index()
print(profitability_stats)
```

```python
volume_stats = df4_filtered.groupby('classification')['Size USD'].agg(['sum', 'mean', 'count']).reset_index()
print(volume_stats)
```

```python
risk_metrics = df4_filtered.groupby('classification').agg({
    'Fee': 'mean',
    'Size USD': 'mean',
    'Start Position': 'mean',
    'Leverage': 'mean'   # Aggregate leverage as mean within groups
}).reset_index()

print(risk_metrics[['classification', 'Fee', 'Leverage']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# Boxplot of Closed PnL by sentiment classification
plt.figure(figsize=(10, 6))
sns.boxplot(x='classification', y='Closed PnL', data=df4_filtered)
plt.title('Profit/Loss Distribution by Market Sentiment')
plt.xticks(rotation=45)
plt.show()
```

```python
# Barplot of average Fee by sentiment classification
plt.figure(figsize=(8, 5))
sns.barplot(x='classification', y='Fee', data=risk_metrics)
plt.title('Average Trading Fee by Market Sentiment')
plt.xticks(rotation=45)
plt.show()
```

```python
# Barplot of average Leverage by sentiment classification (log scale due to wide range)
plt.figure(figsize=(8, 5))
sns.barplot(x='classification', y='Leverage', data=risk_metrics)
plt.yscale('log')
plt.title('Average Leverage by Market Sentiment (Log Scale)')
plt.xticks(rotation=45)
plt.show()
```

```python
daily_metrics = df4_filtered.groupby('date').agg({
    'Closed PnL': 'mean',
    'Size USD': 'sum',
    'Fee': 'mean',
    'Leverage': 'mean',
    'value': 'mean'  # Sentiment score
}).reset_index()
```

```python
correlation_matrix = daily_metrics[['Closed PnL', 'Size USD', 'Fee', 'Leverage', 'value']].corr()
print(correlation_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Daily Trading Metrics and Sentiment Score")
plt.show()
```
